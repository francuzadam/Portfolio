Data Transformation to Visualize

Import the required libraries

```python
import pandas as pd
from datetime import datetime
```

Import the dataset and transform

```python
football_data = pd.read_csv("2021-2022.csv")
football_data["Date"] = pd.to_datetime(football_data["Date"], format='%d/%m/%Y')
football_data["Year"] = football_data["Date"].dt.year
football_data["MonthNum"] = football_data["Date"].dt.month
football_data['Month'] = football_data['Date'].dt.strftime('%B')
football_data["Day"] = football_data["Date"].dt.day
football_data["Result"] = football_data["FTHG"].astype(str) + " - " + football_data["FTAG"].astype(str)
football_data.reset_index(inplace=True)
football_data.head()
```

Export data

```python
football_data.to_excel("input_excel.xlsx")
```
